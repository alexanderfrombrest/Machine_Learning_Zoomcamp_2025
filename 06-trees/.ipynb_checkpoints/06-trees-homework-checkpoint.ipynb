The **DecisionTreeRegressor** (like most machine learning models in scikit-learn) cannot understand text values (e.g., "Apartment" or "Male"). 
It requires all input features (X) to be in a purely numerical format.

Below pipeline is a way to automatically find and convert all categorical text columns into numbers using a method called **one-hot** encoding.

1. **Create List of Dictionaries**
2. **Use DictVectorizer(sparse=False)**
   
   **Why**:
    This tool is the "converter." Its job is to turn that list of dictionaries into a numerical matrix.
    It automatically detects which keys have numerical values (like price) and which have text values (like room_type and city).
    For the text features, it performs one-hot encoding:
    It finds all unique values (e.g., "Apartment", "House", "NY", "SF").
    It creates a new binary column (1 or 0) for each unique value.

4. **Use fit_transform to create X_train**

   **Why**: This is a crucial scikit-learn step that does two things at once:

    * fit() (Learn): The vectorizer "fits" itself to the training data. It scans the entire list of dictionaries to learn the vocabulary. It identifies
   all possible columns (price, room_type, city) and all possible categorical values ("Apartment", "House", "NY", "SF"). It now has a complete "recipe"
   for how to transform the data.
    * transform() (Apply): It uses that learned recipe to convert the data into the final numerical matrix, X_train.

To check categorical feature names we got use **dv.get_feature_names_out()**

##### **Now we train the model**

##### 1. `dt = DecisionTreeRegressor()`

This line of code **initializes** the model.

* **You create an "instance"** of the `DecisionTreeRegressor` class. Think of this as creating a **blank, untrained blueprint** for a decision tree.
* The `dt` object is not yet a tree. It's just an object that "knows" the rules and algorithms for *how* to build one.
* **It loads the default settings (hyperparameters):** Since you didn't specify any settings inside the parentheses, it uses the defaults. The most important one for this process is:
    * **`criterion='squared_error'`**: This is the key. It tells the model that its "job" is to measure its performance by using **Mean Squared Error (MSE)**. This is what it will try to minimize when building the tree.
    * **`max_depth=None`**: This tells the tree it can grow as deep as it needs to. This can lead to overfitting, which is something you'll learn to control later by "tuning" this parameter.

---

##### 2. `dt.fit(X_train, y_train)`

This line **trains** the model.
This is a **recursive** process. Here's how it works, step-by-step:

##### ðŸŒ³ How the Tree is Built

1.  **Start at the Root (All Data):** The algorithm first looks at *all* your training samples (all of `X_train`) in a single "root" node. It calculates the average value of your target (`y_train`) for *all* samples. If it did nothing else, this average would be its one-and-only prediction.

2.  **Find the "Best Split":** This is the most important part.
    * The algorithm iterates through **every single feature** in your `X_train` (e.g., 'horsepower', 'model_year', 'drivetrain=All-wheel drive').
    * For each feature, it tries **every possible split point**.
        * For **continuous features** (like 'horsepower'), it tries splits like `horsepower < 100`, `horsepower < 110`, etc.
        * For **binary features** (like 'drivetrain=All-wheel drive'), the split is simple: `drivetrain=All-wheel drive == 1` (or 0).
    * For *every potential split*, it temporarily divides the data into two groups (e.g., "Group A: horsepower < 110" and "Group B: horsepower >= 110").
    * It then calculates the **Mean Squared Error (MSE)** for Group A and Group B separately, based on their *own* average `y` values.
    * The **"Best Split"** is the *single feature and value* (e.G., `horsepower < 110`) that results in the **lowest combined MSE** for the two new groups. This is the question that does the best job of separating high-value `y`'s from low-value `y`'s.

3.  **Make the Split:** The algorithm "locks in" that best split, creating two new *branches* (or *nodes*).
    

4.  **Recurse:** Now, the algorithm treats each branch as a new, smaller problem.
    * **For Branch A:** It repeats Step 2, looking *only* at the data that ended up in Branch A. It finds the *next* best feature to split on (e.g., maybe `model_year < 2015`).
    * **For Branch B:** It *also* repeats Step 2, looking *only* at its own data. It finds *its* best split (e.g., maybe `drivetrain=All-wheel drive == 1`).

5.  **Stop Splitting (Create "Leaves"):** This process of splitting continues down each branch until a stopping condition is met:
    * A node has reached the **`max_depth`** (if you set one).
    * A node doesn't have enough samples to split (defined by **`min_samples_split`**, which defaults to 2).
    * No further split can be found that significantly reduces the MSE.

6.  **Assign Leaf Values:** When a node stops splitting, it becomes a **"leaf."** The model calculates the **average `y` value** of all the training samples that ended up in that leaf. This average value becomes the **final prediction** for any *new* data point that follows that path down the tree.

---

#### ðŸ The Result

After `dt.fit()` is finished, your `dt` object is no longer a blank blueprint. It is now a **fully trained model**, a complex structure of "if-then" questions and final predictions (the leaf values) based on the patterns it found in your `X_train` and `y_train`.

##### How to Validate a Regression Model

1.  **Prepare Validation Data:** You must apply the *exact same* transformations as you did to the training data.
    * Convert the validation DataFrame to dictionaries: `val_dicts = df_val.to_dict(orient='records')`
    * Use the **.transform()** method of your *already-fitted* `DictVectorizer` (`dv`) to create `X_val`:
        ```python
        X_val = dv.transform(val_dicts)
        ```
2.  **Make Predictions:** Use your trained model (`dt`) to predict on the new `X_val`:
    ```python
    y_pred = dt.predict(X_val)
    ```
3.  **Evaluate Performance (Get a Score):** Compare the model's predictions (`y_pred`) to the real answers (`y_val`).

---

##### Why Use RMSE, Not ROC AUC Score

* **Regression vs. Classification:** You are doing a **regression** problem (predicting a continuous number, like price). `roc_auc_score` is a metric for **classification** problems (predicting a category, like `Yes/No`).
* **Correct Metric:** Use a regression metric like **Root Mean Squared Error (RMSE)**.
* **How to Calculate RMSE:**
    ```python
    from sklearn.metrics import mean_squared_error
    
    # Compare the true values (y_val) to the model's predictions (y_pred)
    rmse = mean_squared_error(y_val, y_pred)
    
    print(f'Validation RMSE: {rmse}')
    ```
* **What RMSE Means:** It's the "average error" of your model, in the same units as your target (`y`). **A lower RMSE is better.**

---

##### Why `.transform()` (and NOT `.fit_transform()`) on Validation Data

* **Why is this critical?** Your trained model (`dt`) expects its input to have the *exact same columns* (features) in the *exact same order* as `X_train`.
* **What Happens if You Use `.fit_transform()` on Validation Data?**
    * **It will crash or give bad results:** The `dv` creates a *new* rulebook based *only* on the validation data. If your validation set (by chance) doesn't have any cars from 'Asia', it won't create the `origin=Asia` column.
    * This new `X_val` matrix will have a different shape (different columns) than `X_train`.
    * Your model (`dt`) will fail, saying it was trained on a different set of features.
* **Rule:** You **only** call `fit` (or `fit_transform`) **once**, on your training data. All other datasets (validation, test, new data) must be processed using **only `.transform()`**.

#### 2. Why Experimenting with Tree Depth is the *Perfect* Next Step

You are 100% correct. Even though your score is low, you **must** experiment with tree depth. Hereâ€™s why:

* **The Problem:** Your current model (`dt = DecisionTreeRegressor()`) was created with all the default settings. The most important default is **`max_depth=None`**.
* **What `max_depth=None` Means:** This tells the tree it can keep splitting and growing as deep as it wants until it perfectly "memorizes" every single data point in the training set.
* **This is called OVERFITTING:** Your current tree is probably massiveâ€”hundreds of levels deep. It has not just learned the *patterns* in your data; it has also learned all the *noise* and *random quirks*.
* **The Trap:** It's possible your validation set is just "lucky" and is very similar to your training set, which is why your RMSE is so low. An overfit model will often fail badly when it sees *truly new* data (i.e., your test set).

By experimenting with `max_depth`, you are performing **hyperparameter tuning** to **regularize** (or simplify) the model.

* **Your Goal:** You want to find the "sweet spot" for `max_depth`.
    * A tree that is **too deep** (like your current one) **overfits**.
    * A tree that is **too shallow** (e.g., `max_depth=1`) is too simple and **underfits** (it can't even learn the main patterns).

* **The Process:** You should now try training and validating your model in a loop, testing different values for `max_depth` (e.g., 1, 2, 3, 5, 10, 20) and plotting the RMSE for each. The `max_depth` that gives you the **lowest validation RMSE** is your best, most general-purpose model.

#### 3. Confirming Overfitting
Just like a `roc_auc_score` of 1.0 on training data signals perfect "memorization" (overfitting) in classification, an **RMSE of 0.0 (or very close to it) on training data signals the exact same thing for regression.**

#### Why Random Forest is a Great Next Step

A Random Forest is designed to fix this exact problem.

1.  **It's an Ensemble:** It works by building *many* decision trees (e.g., 100 of them) in parallel.
2.  **It Uses Randomness:**
    * Each tree is trained on a *random sample* of your data (this is called "bagging").
    * When splitting a node, each tree is only allowed to look at a *random subset* of your features (e.g., it might only be allowed to pick from `horsepower`, `origin=USA`, and `num_cylinders`).
3.  **It Averages the Results:** To make a final prediction, the Random Forest asks all 100 of its trees, "What do you predict?" and then **averages their answers**.

This process is incredibly powerful. It combines hundreds of different "weak" or "simple" trees, each of which has learned a small part of the pattern, into a single, strong model that generalizes extremely well and is much less prone to overfitting than a single, deep decision tree.

You will now need to tune new hyperparameters, with the most important one being **`n_estimators`** (the number of trees to build).


#### How finding the best split algorithm works:

    for f in features:
        find all thresholds for f
    
        for t in *thresholds*:
            1. split dataset using "feature > threshold" condition
            2. compute the impurity of this split        
1. **Select the condition with the lowest impurity** - find the best split
2. **Stop if max_depth is reached**
3. **If left is sufficiently large and not pure --> repeat for the left**
4. **If right is sufficiently large and not pure --> repeat for the right**



#### Stopping criteria: how do we know when to stop growing decision tree?
* Group is already pure. The split does not make sense anymore.
* Tree reached depth limit.
* Group is too small to split it, ex. we can set that if group is < 5 values we do not split it. We can set mininum_size of the group.



#### Decision Tree parameter tuning
Usually, for large datasets, where we can't test all combinations of parameters, we test:
    1. max_depth
    2. min_samples_leaf

````python
def parse_xgb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_auc', 'val_auc']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
````
